```python
import os
import pandas as pd
import re
```

```python
def read_csv_files(directory):
    all_dataframes = []  # Initialize an empty list to store dataframes

    # Loop through each file in the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('bills.csv'):  # Check if the file is a CSV file
                file_path = os.path.join(root, file)  # Get the full file path
                df = pd.read_csv(file_path)  # Read the CSV file into a pandas dataframe
                all_dataframes.append(df)  # Append the dataframe to the list

    # Concatenate all dataframes into a single dataframe
    combined_df = pd.concat(all_dataframes, ignore_index=True)

    return combined_df

# Set the directory path where your CSV files are located
directory_path = 'unstructured_data'

# Call the function to read CSV files recursively and concatenate them
result_df = read_csv_files(directory_path)

# Print the combined dataframe
print(result_df.shape)
```

```python
result_df['State'] = result_df['url'].apply(lambda x: re.search(r'/([A-Z]{2})/bill/', x).group(1))
```

```python
def read_csv_files(directory, quotechar='"'):
    all_dataframes = []  # Initialize an empty list to store dataframes

    # Loop through each file in the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('documents.csv'):  # Check if the file is a CSV file
                file_path = os.path.join(root, file)  # Get the full file path
                df = pd.read_csv(file_path)  # Read the CSV file into a pandas dataframe
                all_dataframes.append(df)  # Append the dataframe to the list

    # Concatenate all dataframes into a single dataframe
    combined_df = pd.concat(all_dataframes, ignore_index=True)

    return combined_df

# Set the directory path where your CSV files are located
directory_path = 'data'

# Call the function to read CSV files recursively and concatenate them
documents_result_df = read_csv_files(directory_path)

# Print the combined dataframe
print(documents_result_df.shape)
```
